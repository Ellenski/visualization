```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Create dataset to visualize

np.random.seed(613)
x = np.arange(50) #Range from 0 - 49
y1 = np.random.randint(0, 100, 50) #list of 50 random integers ranging from 0 to 99
y2 = np.random.randint(0, 100, 50) #Same as y1
```

```python
#Make a line plot of both y variables on same axis

fig, ax = plt.subplots(figsize = (5,3)) #initialize plot
ax.plot(x, y1) #First line plot with y1
ax.plot(x, y2) #Second line plot with y2
fig.show() #Show the plot, not necessary for VS code
```

```python
#Adding a legend to the plot

fig, ax = plt.subplots(figsize = (5,3)) 
ax.plot(x, y1, label = "Person 1") #Specify name of this line with label
ax.plot(x, y2, label = "Person 2") #Specify name of this line with label
ax.legend(loc = "lower right") #Add in legend so it appears in the lower right corner of the graph using "loc"
fig.show() 
```

```python
#Modify the legend 

fig, ax = plt.subplots(figsize = (5,3)) 
ax.plot(x, y1, label = "Person 1") 
ax.plot(x, y2, label = "Person 2") 
ax.legend(loc = "lower right",
          frameon = True, #Add frame around the legened
          fontsize = 12, #control the font size in the legend
          ncol = 2, #Indicate the number of columns in the legends --> here one column for each "Person"
          shadow = True) #Add a drop shadow to the legend = EW! BAD!
fig.show() 
```

```python
#Move the legend outside the plot

fig, ax = plt.subplots(figsize = (5,3)) 
ax.plot(x, y1, label = "Person 1") 
ax.plot(x, y2, label = "Person 2") 
ax.legend(loc = "upper left", #Indicate the position of the legend is upper left
          bbox_to_anchor = (1,1)) #Will indicate the legend should appear outside of plot and the coordinates of where it should be, use upper right corner of legend as the anchor point and then move it with the coordinate = 100 % y axis and 100% x axis
fig.show() 
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") #indicate scatter plot
ax.scatter(x, y2, label = "Person 2") #indicate scatter plot
ax.legend(loc = "lower right")

#Add a text annotation to the graph
ax.text(10, 95, "This value is important!") #Indicate x value, y value and then message
fig.show
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") #indicate scatter plot
ax.scatter(x, y2, label = "Person 2") #indicate scatter plot
ax.legend(loc = "lower right")

#modify a text annotation to the graph
ax.text(10, 95, "This value is important!",
        ha = 'center', #Indicates how to align text
        color = 'red', #Indicates the color of the text
        size = 20) #indicates the font size
fig.show

#This looks pretty terrible :(
```

```python
#Make simple empty set of axes
fig, ax = plt.subplots()
ax.axis([0,10,0,10]) #x and y axis limits

#Add in text annotations in three different ways
ax.text(1,5, ". Data:(1,5)", transform = ax.transData) #based on data --> values in axis
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes) #Based on axes --> percentage of axis
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure) #Based on Figure
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")

#Add arrow in addition to text
ax.annotate('This is important!', #Text annotation that we want
            xy = (10, 95), #x and y location of the arrow
            xytext = (20, 94), #x and y of where we want text to go
            arrowprops = dict(facecolor = 'black', shrink = 0.05)) #Arrow details, color within arrow outline and how much gap between arrow and text

fig.show()
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")

#Modify arrow in addition to text
ax.annotate('This is important!', 
            xy = (10, 95), 
            xytext = (20, 94), 
            arrowprops = dict(arrowstyle = 'wedge', #Change the arrow style
                              color = 'hotpink'))  #Chage arrow color

fig.show()
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")

#Remove the axis ticks and labels
ax.yaxis.set_major_locator(plt.NullLocator()) #Remove tick marks and y axis label for y axis = Null
ax.xaxis.set_major_formatter(plt.NullFormatter()) #Remove only the axis label for x axis = Null
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")
#limit the number of axis ticks on the x axis
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #Limit to only 3 ticks on the x axis, ie designste and max number of tick marks
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")
#indicate intervale for x axis tick marks
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #Tick mark every 5 x axis values
```

```python
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")
#Rotate the x axis labels
plt.xticks(rotation = 45, ha = 'right') #Rotate x axis labels 45 degrees, align them to the right of tick mark
```

```python
#Font dictionary indicating font family and font color
font1 = {'family':'serif', 'color':'indigo'}
#Plot data as scatter point
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc = "lower right")
#Modify x axis title font
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1) #Indicate title, size of font and the font attributes based on created font dictionary
```
