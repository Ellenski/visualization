```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

#Sample data
np.random.seed(613) #Use a random seed
x1 = np.arange(50) #range of data from 0 to 49
y1 = np.random.randint(0, 75, 50) #range of 50 values from 0 to 74
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) #ONE PIECE :D
y2 = np.array([110, 180, 240, 99, 220]) #Five values, one for each name

```

```python
#We want two plot next to each other --> define two sets of axes

fig,(ax1, ax2) = plt.subplots(ncols =2, #Defining two sets of axes as subplots shown in two columns
                             nrows = 1, #Show in one row
                             figsize = (7,3)) #Wider space to fill because there are two graphs
```

```python
#Add in the data

fig,(ax1, ax2) = plt.subplots(ncols =2, 
                             nrows = 1, 
                             figsize = (7,3)) 

ax1.scatter(x1,y1)#We want first graph to be a scatter plot
ax2.bar(x2,y2) #We want second graph to be a bar plot
fig.show()
```

```python
#Add modifications to the plots

ig,(ax1, ax2) = plt.subplots(ncols =2, 
                             nrows = 1, 
                             figsize = (7,3)) 

ax1.scatter(x1,y1, marker = '*', color = 'orange')
ax1.set_title('Starry Night')
ax2.barh(x2,y2, color = 'purple') 
ax2.set_title('One Piece')
fig.show()
```

```python
#Anoter way to plot subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']], ##Think of this like tiles --> this is what indicates how much of the space the different axes will be taking up
                                    figsize = (7,4))
for label, ax in someaxes.items():
    ax.text(0.5, 0.5, f'This is {label!r}',
            fontsize = 14,
            ha = 'center',
            transform = ax.transAxes) #Label the different ax
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], #someaxes is not a placeholder --> must be used this way
                                    ['ax2', 'ax3']],
                                    figsize = (7,4))
someaxes["ax1"].scatter(x1,y1) #add the data to the axes --> scatter plot here
someaxes["ax2"].bar(x2,y2) # bar plot
someaxes["ax3"].plot(x1,y1) #line plot
plt.show()
```

```python
#Modifying plot mosaic
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']],
                                    figsize = (7,4))
someaxes["ax1"].scatter(x1,y1) 
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1) 
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18) #This one is completely cut off --> too big
someaxes["ax2"].set_xlabel('Another Label',fontsize=18) #This one is fine
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 ) #This one is partially cut off --> also too big

plt.show()
```

```python
#Use layouts to make sure all elements are shown, not squished or cutoff
#Modifying plot mosaic
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = "constrained") #Add layout
someaxes["ax1"].scatter(x1,y1) 
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1) 
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18) 
someaxes["ax2"].set_xlabel('Another Label',fontsize=18) 
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 ) 

plt.show()

#Now everything fits properly
```

```python
#Multiple viz on one axes object

x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes 
fig, ax = plt.subplots(figsize=(7, 3))
#call bar and plot elements to the same axes 
ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")

#Two grpahs on the same axes
```

```python
#Add error information
#Calculate standard deviation for y2
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = "red")
#Add in the error bars
ax.errorbar(x, 
            y2, 
            yerr = y2_sd, #The error information that we calculated
            fmt = "none") #no format
```

```python
#Customizing errorbar appearance

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, 
        color = "red")
ax.errorbar(x, 
            y2, 
            yerr = y2_sd, 
            fmt = "none", 
            ecolor= "indigo", #Color of error bar
            elinewidth= 4, #with of line of error bar
            capsize = 6, #width of the cap on error bar
            capthick= 4 )# thickness of cap
```

```python
#Don't wish to see error bars for every value

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, 
        color = "red")
ax.errorbar(x, 
            y2, 
            yerr = y2_sd, 
            fmt = "none", 
            ecolor= "indigo", 
            elinewidth= 4, 
            capsize = 6, 
            capthick= 4,
            errorevery =2) #errorbar every two points
```

```python

#Import specific necessary commands and packages
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
#Get, save and open image
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
image
```

```python
#Getting ready to add image to graph
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = "red")

#create ax where we can put image we downloaded
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                         0.11, # y coordinate (ON FIGURE, NOT AXES)
                         0.15, # image width
                         0.35]) # image height

#Added in empty axes object --> see where image is going to go 
```

```python
#Now add image in

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1, 
                         0.11, 
                         0.15, 
                         0.35]) 

ax_image.imshow(image) #Call the saved image
ax_image.axis('off') #removed lines/box around the image
plt.show()
```
