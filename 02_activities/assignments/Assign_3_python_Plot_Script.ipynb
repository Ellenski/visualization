```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO

```

```python
moose_data = pd.read_csv('Moose_Population_Data_Processed.csv')
moose_data
```

```python
#Making Figure
font1 = {'family' : 'sans-serif', 'color' : 'black', 'size' : 22, 'fontweight': 'bold'}
font2 = {'family' : 'sans-serif', 'color' : 'black', 'size' : 16, 'fontweight': 'normal'}

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(moose_data["SurveyYear"],
        moose_data["Sum_Obsv_Moose"],
        color = 'cornflowerblue',
        linewidth = 1,
        marker = ".",
        markeredgecolor = 'blue', 
        markerfacecolor = 'blue'
        )

ax.set_title('Ontatio Moose Population Per Year', fontdict = font1) 
ax.set_ylabel('Total Moose Population', fontdict = font2) 
ax.set_xlabel('Year', fontdict = font2) 
plt.xticks(np.arange(1975, 2026, step=5), rotation = 75)
ax.tick_params(axis='both', which='major', labelsize=11)

arr_img = plt.imread("Moose.png")

ax_image = fig.add_axes([0.1, 
                         0.58, 
                         0.20, 
                         0.3]) 
ax_image.imshow(arr_img) #Call the saved image
ax_image.axis('off')

plt.savefig('Assign_3_python_Plot.png', bbox_inches='tight')
```
